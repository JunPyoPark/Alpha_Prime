```python
from googletrans import Translator
translator = Translator()
```

```python
import pandas as pd
```

```python
stock_list = ['000270','003490','035720','051910','068270']
```

```python
for code in stock_list:
    en = []
    data = pd.read_csv('labeled_with_content'+code+'.csv',index_col=0)
    c = 0
    for idx,row in data.iterrows():
        try:
            en.append(translator.translate(row['title'], dest='en').text)
            c = c + 1
        except:
            en.append(' ')
        print(c,'/',len(data),' translated to English')
    
    data['en_title'] = en
    data.to_csv('translated_'+code+'.csv')
    print(code)
```
