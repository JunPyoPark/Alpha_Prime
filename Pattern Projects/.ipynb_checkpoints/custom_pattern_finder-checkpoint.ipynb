사용자가 원하는 패턴을 미리 만들어 놓고 그와 유사한 패턴을 찾아 볼 수 있는 모듈

### get_w_pattern : w 모양(쌍바닥) 패턴을 만들어주는 함수

#### idx: 인덱스, percent : W 패턴의 몇 % 나 생성 할 것인지 결정하는 value

> eg. length(idx) = 100 이고 percent=0.5 라면 길이 100 짜리 V 모양 패턴이 생성됨

```python
def get_w_pattern(idx, percent=1):
    length = len(idx)
    extended_length = int(length // percent)
    w = pd.Series([np.NaN] * extended_length)
    ### 아래 값들은 상승형 W(쌍바닥) 패턴이 되도록 임의로 정한 값임 ###
    ### 추후 ML, bootstrap 기법등을 사용하여 고도화 예정 ###
    w.iloc[0] = 0.75
    w.iloc[extended_length//4] = 0
    w.iloc[extended_length//2] = 0.75
    w.iloc[extended_length*3//4] = 0.25
    w.iloc[-1] = 1
    w = w.interpolate().iloc[:length]
    w.index = idx
    return w
```

```python
w = get_w_pattern(range(100),0.5)
w.plot(figsize=(10,7))
plt.show()
```

```python
w = get_w_pattern(range(10),0.9)
w.plot(figsize=(10,7))
plt.show()
```

```python
test_data = close.iloc[-35:-15].T
```

## Min-Max Scaling 함수 만들기

min-max scaling 할 때 변동폭이 큰 녀석들만 선별함

```python
def get_minmax_scale(series):
    diff = series.max() - series.min()
    if diff / series.mean() > 0.15: # 변동폭이 15% 이상 되는 것만 선별
        return (series-series.min())/diff
    else:
        return [np.NaN] * len(series)
```

```python
test_data = test_data.apply(get_minmax_scale, axis=1) # 종가 테이블의 최대,최소를 1,0 으로 통일(Normalizing)
test_data = test_data.dropna(axis=0) # blank 제거
print(len(test_data))
```

```python
test_data.T.index
```

```python
w = get_w_pattern(test_data.T.index,1)
# w.plot(figsize=(10,7))
# plt.show()
```

```python
def get_cos_similar(target_data, test_data) :    
    cos_similar = pd.Series()
    for i in test_data.index :
           cos_similar = cos_similar.append(pd.Series(1 - spatial.distance.cosine(target_data, test_data.loc[i]), index = [i]))
    cos_similar = cos_similar.sort_values(ascending=False) # 유사도 높은 순서로 정렬
    return cos_similar
```

```python
cos_similar = get_cos_similar(w, test_data)
```

```python
def plt_cosine_result(cos_similar):
    plt.figure(figsize=(12,7))
    plt.plot(w, label = 'Target Chart')
    plt.plot(test_data.loc[cos_similar.index[0]], label = 'Most Similar Chart')
    plt.plot(test_data.loc[cos_similar.index[1]], label = 'Second Similar Chart')
    plt.plot(test_data.loc[cos_similar.index[2]], label = 'Third Similar Chart')
    plt.plot(test_data.loc[cos_similar.index[3]], label = 'Fourth Similar Chart')
    plt.legend()
    plt.show()
    print('Most Similar Code is : ', cos_similar.index[0], " -> ", cos_similar[0])
    print('Second Similar Code is : ', cos_similar.index[1], " -> ", cos_similar[1])
    print('Third Similar Code is : ', cos_similar.index[2], " -> ", cos_similar[2])
    print('Fourth Similar Code is : ', cos_similar.index[3], " -> ", cos_similar[3])
```

```python
plt_cosine_result(cos_similar)
```

## PIP(Perceptually Important Point) Algorithm

```python
def get_pips(series, point_number):
    series.index = range(len(series.index))
    conversion_list = series.reset_index().values.tolist()
    pip_list = pip(conversion_list, point_number)
    tmp_array = np.array(pip_list)
    return pd.Series(tmp_array[:,1], index=tmp_array[:,0])
```

point_number : PIP 알고리즘으로 뽑아낼 point 의 수

```python
pip_df = test_data.apply(get_pips, axis = 1, args = (5,)) # 여기서는 important points 를 7개 뽑는 걸로 실험
```

```python
pip_df
```

linear 하게 NaN 값들이 채워지는 것 확인!

```python
pip_df = pip_df.interpolate(axis=1)
```

```python
def plt_cosine_result(cos_similar):
    plt.figure(figsize=(12,7))
    plt.plot(list(w), label = 'Target Chart')
    plt.plot(pip_df.loc[cos_similar.index[0]], label = 'Most Similar Chart')
    plt.plot(pip_df.loc[cos_similar.index[1]], label = 'Second Similar Chart')
    plt.plot(pip_df.loc[cos_similar.index[2]], label = 'Third Similar Chart')
    plt.plot(pip_df.loc[cos_similar.index[3]], label = 'Fourth Similar Chart')
    plt.legend()
    plt.show()
    print('Most Similar Code is : ', cos_similar.index[0], " -> ", cos_similar[0])
    print('Second Similar Code is : ', cos_similar.index[1], " -> ", cos_similar[1])
    print('Third Similar Code is : ', cos_similar.index[2], " -> ", cos_similar[2])
    print('Fourth Similar Code is : ', cos_similar.index[3], " -> ", cos_similar[3])
```

```python
cos_similar = get_cos_similar(w, pip_df) # pip 전처리된 데이터를 사용하여 유사도 비교
plt_cosine_result(cos_similar) # 시각화 결과 출력
```
