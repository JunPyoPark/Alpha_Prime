# Pattern Dicision Helper

과거의 패턴에서 현재 차트의 흐름을 발견하여 그 이후 추이를 살펴 볼 수 있는 모듈

```python
import numpy as np
import pandas as pd
from scipy import spatial
import matplotlib.pyplot as plt
%matplotlib inline
from jupyterthemes import jtplot
import datetime
jtplot.style(theme='grade3')
```

## 주가 데이터 읽기

```python
close = pd.read_csv('C:\\Data\\Abnormal Price Separated Table.csv',index_col = ['Date'])
close.index = pd.to_datetime(close.index, format = '%Y-%m-%d')
close
```

증자나 액면 분할 등으로 주가가 갑자기 변한 경우는 현재 고려 안함 <br>

```python
class find_pattern():
    
    def __init__(self,code,pattern_period,after_period,ranking):
        self.code = code
        self.pattern_period = pattern_period
        self.after_period = after_period
        self.ranking = ranking
        
        self.target_data = close[self.code].tail(self.pattern_period) #목표 패턴 설정, 현재 기준 일정 기간 이전 부터의 데이터
        self.target_data = get_minmax_scale(self.target_data) # 0~1 스케일링
        
        data = close[code][:-pattern_period-after_period] # 기간 동안 rolling 을 돌릴 데이터, 현재 선택한 종목과 같은 종목만 사용
        cos_result = data.rolling(window=pattern_period).apply(lambda x : self.get_cos_similar(x)) # 예재에서는 cos measure 적용
        self.result = cos_result.sort_values(ascending=False)
            
    def get_minmax_scale(self,array_like): # min_max scaling
        diff = array_like.max() - array_like.min()
        return (array_like-array_like.min())/diff
    
    def get_cos_similar(self,array): # cosine similar measure
        array = self.get_minmax_scale(array)
        return 1 - spatial.distance.cosine(self.target_data, array)
    
    def own_measure(self,array): # you can add your own measure / eg. euclidean, manhatten, correlation...
        pass
    
    def plot_result(self): # plot result
        plt.figure(figsize=(12,7))
        plt.plot(np.array(self.target_data),color='red')
        count = 0
        sensitivity = 5 # 같은 구간이 하루/이틀 평행이동 되어 나오는 것을 방지하기 위한 민감도
        np_date = np.array([np.datetime64('1900-01-01')]*self.ranking)
        for i in self.result.index: # 유사도 상위 랭킹부터 for 문을 실행
            # tolerance 를 설정함, 임의로 두가지 방식을 만듬
            tolerance = int(self.pattern_period/sensitivity) # 구간 시작일이 패턴기간/민감도 이상 차이나는 결과만 선택
            # tolerance = int(np.sqrt(self.pattern_period))  # 구간 시작일이 sqrt(패턴기간) 이상 차이나는 결과만 선택
            if abs(np_date - np.datetime64(i)).min() > np.timedelta64(tolerance,'D') : # tolerance 이상의 결과만 선택
                iloc = close[self.code].index.get_loc(i)
                similar_data = np.array(close[self.code][iloc-self.pattern_period:iloc+self.after_period]) # 유사 패턴과 그 이후 추이 출력
                plt.plot(self.get_minmax_scale(similar_data),color='gray',linestyle='dashed')
                np_date[count] = i
                if count > self.ranking-2:
                    break
                else:
                    count=count+1
        # 결과와 구간 종료일 데이터 출력
        plt.show()
        print(np.sort(np_date))
```

```python
# 현재일 기준으로 20일 이전부터의 패턴을 기준으로 전 구간동안 그와 유사한 패턴을 5개 뽑아 그 패턴들과 이후 10일간 움직임을 출력
a = find_pattern('A090430',20,10,5)
```

```python
a.plot_result()
```
