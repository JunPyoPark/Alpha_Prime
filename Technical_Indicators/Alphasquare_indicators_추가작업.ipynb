## SMA : Simple Moving Average (단순이동평균)

기술적분석의 기초 지표로 주가의 흐름에서 들쑥날쑥한 부분, 잡음(noise)를 제거하고 추세를 파악하려는 방법으로 사용된다. 

이동평균선은 투자자에 따라서 다양한 기간으로 산출할 수 있다. 대체로 5일, 20일, 60일, 120일, 240일 등의 이동평균이 보편적으로 사용된다.

아울러 이동평균선은 지지선/저항선으로서의 의미를 가진다. 이동평균은 과거 특정한 기간의 가격을 평균한 것이다. 따라서 그 기간 동안의 가격을 대표하는 값이며 이동평균 안에는 그동안의 가격 움직임이 녹아있다.

이동평균을 산출하는 공식은 다음과 같다.

###  $$n일\,\,이동평균=\frac{P_0+P_{-1}...+P_{-(n-1)}}{n}$$

### $$P_0 = 당일의 \,\,시장가격$$

### $$P_{-n} = n일\,\, 전의 \,\,시장가격$$

#### 매매방법

골든크로스 : 단기이동평균선이 장기이동평균선을 상향돌파 -> 매수 <br>
데드크로스 : 단기이동평균선이 장기이동평균선을 하향돌파 -> 매도

지지선과 저항선을 이용한 매매방법 <br>
주가가 하락하여 이동평균에 근접할 경우, 이동평균이 지지선이 되는 것을 확인하고 매입하고, 반대로 주가가 상승하여 이동평균이 근접할 경우, 이동평균이 저항선이 되는 것을 확인하고 매도하는 방식

```python
def SMA(close, timeperiod=5):  # timeperiod = n일간 이평선 결정
    return close.rolling(window=timeperiod).mean()
```

```python
SMA(close,20).plot()
```

## EMA : Exponential Moving Average (지수이동평균)

지수이동평균(Exponential Moving Average)은 과거의 모든 기간을 계산대상으로 하며 최근의 데이타에 더 높은 가중치를 두는 일종의 가중이동평균법이다.
단순이동평균의 계산법에 비하여 원리가 복잡해 보이지만 실제로 이동평균을 산출하는 방법은 전일의 지수이동평균값과 평활계수(smoothing constant) 그리고 당일의 가격만으로 구할 수 있으므로 전일의 지수이동평균값만 구해진다면 오히려 간단한 편이다.
따라서 지수이동평균은 단순이동평균에 비해 몇가지 중요한 강점을 가진다.
첫째는 가장 최근의 일자에 가장 큰 가중치를 둠으로 해서 최근의 시장분위기를 잘 반영한다는 점이고, 둘째는 단순이동평균에서와 같이 오래된 데이타를 갑자기 제외하지 않고 천천히 그 영향력을 사라지게 한다는 점이다.
또한 전 기간의 데이타를 분석대상으로 함으로써 가중이동평균에서 문제되는 특정 기간의 데이타만을 분석대상으로 한다는 단점도 보완하고 있다.

지수이동평균을 산출하는 공식은 다음과 같다.

###  $$EMA_{today}=(Price_{today} \times K) + (EMA_{yesterday} \times (1-K))$$

### $$K=\frac{2}{N+1}\,\,where \,\,N \,\,is\,\, the\,\, Period$$

### $$Price_{today} = 오늘 종가$$ <br>
### $$EMA_{yesterday} = 어제의 EMA 값$$ <br>
### $$EMA_{today} = 오늘 EMA 값$$ <br>

#### 매매방법

골든크로스 : 단기이동평균선이 장기이동평균선을 상향돌파 -> 매수 <br>
데드크로스 : 단기이동평균선이 장기이동평균선을 하향돌파 -> 매도

지지선과 저항선을 이용한 매매방법 <br>
주가가 하락하여 지수이동평균에 근접할 경우, 지수이동평균이 지지선이 되는 것을 확인하고 매입하고, 반대로 주가가 상승하여 지수이동평균이 근접할 경우, 지수이동평균이 저항선이 되는 것을 확인하고 매도하는 방식

```python
def EMA(close, timeperiod=30):  # Exponential Moving Average : 지수이동평균
    k = 2/(1+timeperiod) # k : smoothing constant
    close = close.dropna()
    ema = pd.Series(index=close.index)
    ema[timeperiod-1] = close.iloc[0:timeperiod].sum() / timeperiod
    for i in range(timeperiod,len(close)):
        ema[i] = close[i]*k + ema[i-1] * (1-k)
    return ema
```

```python
 EMA(close,20).plot()
```

## BBANDS : Bollinger Band (볼린저 밴드)

통계학에 따르면 표본의 숫자가 어느 수준 이상으로 늘어날 경우 그것의 분포는 대체로 정규분포(Normal distribution)로 나타난다.정규분포에서는 어떤 변수가 평균을 중심으로 +2배 표준편차 이상이 될 가능성은 2.25%에 불과하며, 같은 논리로 어떤 변수가 평균을 중심으로 -2배 표준편차 이하가 될 가능성도 역시 2.25%에 불과하게 된다. 볼린저밴드는 바로 이런 점에 착안하였다.

볼린저밴드는 중간 밴드, 위쪽 밴드 그리고 아래쪽 밴드의 3개 밴드로 구성된다. 중간 밴드는 추세의 중심역할을 하는데, 통상 20일 이동평균선이 사용된다(물론 분석가의 재량에 따라 다른 기간의 이동평균선을 사용할 수도 있다). 그리고 위쪽 밴드는 중간 밴드에다 표준편차 2배를 더한 값으로 나타내며, 아래쪽 밴드는 중간 밴드에다 표준편차 2배를 차감한 값으로 나타낸다.

볼린저 밴드는 다음의 공식으로 산출된다.

### $$중간\,\,밴드=20일 \,\,단순이동평균선$$
### $$상단\,\,밴드=중간\,\,밴드 + 2 \times (20일\,\,표준편차)$$
### $$하단\,\,밴드=중간\,\,밴드 - 2 \times (20일\,\,표준편차)$$

#### 매매방법

볼린저 밴드를 활용한 여러가지 매매방법이 존재하며 여기서는 가장 간단한 매매방법만 살펴보도록 한다.

주가가 상승하여 위쪽 밴드에 근접할 경우, 우리는 통계적으로 주가가 위쪽 밴드를 넘어설 확률이 2.25%에 불과하다는 것을 안다. 따라서 주가는 위쪽 밴드를 넘어서지 못할 가능성이 높다. 다시 말하여 위쪽 밴드가 저항선으로 작용할 것이다. 그러기에 주가가 위쪽 밴드에 근접할 때가 매도 타이밍이 된다. 그 반대의 경우도 성립한다. 주가가 하락하여 아래쪽 밴드에 근접할 경우, 통계적으로 보아 주가는 아래쪽 밴드를 넘어서지 못할 공산이 매우 높다. 아래쪽 밴드가 지지선이 될 것이므로 주가가 아래쪽 밴드에 근접할 때가 매수 타이밍이 된다.

밴드의 폭을 살피는 방법도 고려할 수 있다. 밴드의 폭이 좁아지고 있다면 그동안의 변동성이 낮았다는 의미가 된다. 조만간 큰 폭의 가격변동이 나타날 것을 예고하는 신호이다. 다만 볼린저밴드로는 '변동성'이 커질 것은 예측할 수 있어도 '방향'은 예측할 수 없다.

### Params


close(종가) : pandas.Series <br>
timeperiod(기간) : integer, default = 20 <br>
nbdevup = nbdevdn (승수) : integer, default = 2

```python
def BBANDS(close, timeperiod=20, nbdevup=2, nbdevdn=2, matype=0): # 볼린저 밴드  
    middle = SMA(close, timeperiod)
    std = close.rolling(window=timeperiod).std()
    upper = middle + nbdevup * std
    lower = middle - nbdevdn * std
    df = pd.concat([upper, middle, lower],axis=1)
    df.columns = ['upper','middle','lower']
    return df
```

```python
jp_bbands = BBANDS(close)
jp_bbands.plot(figsize=(12,7))
plt.show()
```

매물대 차트는 기존의 캔들차트나 또는 바차트에다 최근까지의 거래량을 가격대별로 나타낸 것이다. 그러면 어떤 주가 수준이었을 때 거래가 얼마나 많았는지를 쉽게 알 수 있다. 일반적으로 매물대차트는 차트가 그려지는 기간의 주가를 10~12단계로 나누고, 각각의 거래량을 차트에 중복하여 그리는 방법으로 나타낸다.

거래가 많았다는 것은 그 구간에서 매도한 사람도 많았으며 동시에 매입한 사람도 많았다는 뜻이다. 그들 중에서 상당수는 매입한 가격보다 현재의 가격이 하락하여 손해를 보고 있는 상태일 터, 따라서 현재의 주가가 반등하여 그 구간에 근접할수록 본전을 찾으려는 매물이 쏟아질 것이라고 쉽게 예상할 수 있다.그 구간이 바로 매물대이다. 결국 매물대는 '저항선'의 역할을 한다.

매물대는 화면에 표시된 차트상의 기간에 따라 달라진다 즉 산출기간에 따라 매물대가 들쑥날쑥 해진다. 따라서 하나의 차트로는 정확한 매물대를 파악할 수 없다는 약점이 있다.

## PSAR : 파라볼릭

추세전환을 파악하는 데에 효과적인 지표이다. 특히 손절(Stop Loss)을 어느 수준에 두어야 할지 모르는 투자자들에게는 매우 유용하다. 다만 다소 느리다는 것이 약점이다.

PSAR을 산출하는 공식은 다음과 같다.

### $$PSAR_{today} = PSAR_{yesterday}+AF \times (EP - PSAR_{yesterday})$$

### $$AF=Acceleration Factor$$ 

### $$EP=Extreme Point$$ 

AF(Acceleration Factor), 가속요소는 0.02 부터 시작하여 하루가 지날 때마다 매일 0.02가 추가된다. 하지만 무한정 커지지는 않으며 0.20이 최대이다. 따라서 AF가 일단 0.20 수준에 이르는 경우면 그 이후에는 계속 0.20을 사용한다.

EP(Extreme Point), 즉 극단점은 추세가 바뀌는 가격수준으로 정해진다. 예컨대 추세가 상승세에서 하락세로 바뀔 때라면 EP는 직전 최고가격이 되고, 반대로 추세가 하락세에서 상승세로 바뀔 때라면 EP는 직전 최저가격이 된다.

이때 추세가 막 바뀌는 첫 번재 날의 PSAR 은 EP와 같은 값으로 설정된다.

### Parameters

increment: Number (0.02), 가속증가량

initialAccelerationFactor: Number (0.02), 가속요소

maxAccelerationFactor: Number(0.2), 최대가속요소

### 매매방법

PSAR은 추세전환 시점을 알려주므로 주가와 PSAR이 닿을 때가 매매 시기이다. 

주가가 상승하다가 더 오르지 못하고 하락하여 아래쪽에서 포물선을 그리며 상승하는 PSAR을 건드리면 매도 시점

주가가 하락하다가 더 하락하지 않고 상승하여 위쪽에서 포물선을 그리면서 하락하고 있는 PSAR을 건드리면 매수 시점

PSAR을 통해 매매 시기를 잡기보다 보통의 경우 손절(Stop Loss)를 결정하는 데에 사용된다.

즉, 상승추세로 판단하여 주식을 매입하였을 경우 주가가 더 오르지 못하고 PSAR을 건드리면 즉각 손절하는 것이 바람직하다.

# 엔벨로프(price envelopes)

주가가 이동평균선을 중심으로 일정한 밴드를 형성한다는 개념에서 출발하였다. 지지선이나 저항선의 위치를 파악하기 용이하다. 다만 주가가 엔벨로프를 돌파할지의 여부를 알 수 없다는 단점이 있다.

볼린저밴드의 경우 밴드의 폭은 '표준편차'로 정해진다. 따라서 밴드의 폭은 들쑥 날쑥하다. 그러나 엔벨로프의 경우 중간밴드를 중심으로 미리 정해진 '일정한 폭'으로 범위가 정해지므로 폭이 일정하다.

밴드 폭의 경우 분석가에 재량에 따라 달라지며 차트를 작성하는 기간이 단기가 될수록 엔빌로프의 폭이 좁아져야 한다. 반대로 차트를 작성하는 기간이 장기가 될수록 엔빌로프의 폭은 넓어져야 한다. 기간이 길어질수록 주가의 변동성은 늘어날 수밖에 없기 때문이다.

### 매매방법

기본적으로 엔벨로프는 그 자체가 지지선 또는 저항선의 역할을 하는 것으로 간주되므로 볼린저 밴드와 같이 주가가 엔빌로프에 근접할 때가 매매의 타이밍이 된다. 주가가 지지선 근처에 근접하면 의당 매입기회로 삼고 거꾸로 주가가 저항선 근처에 접근하면 매도기회로 삼아야 하는 것은 마땅하다.

엔빌로프는 종종 무너지기도 하는데, 이럴 때라면 즉각 손절하거나 또는 포지션의 방향을 바꾸는 순발력이 요구된다. 구체적으로 말하여 주가가 상승하여 저항선으로 간주되던 위쪽 밴드마저 돌파하였다면 기존의 매입세와 매도세 간의 균형은 깨지고 다시 새로운 균형이 만들어질 것이며, 여기에서는 매입세가 훨씬 더 강력한 힘을 가지는 것으로 해석된다. 결국 상승세가 훨씬 강화되는 것이므로 추격매입이라도 단행해야 한다. 반대로 지지선의 역할을 하던 아래쪽 밴드가 무너진다면 매도세가 훨씬 더 새롭게 강화되는 것으로 해석되므로 즉각 추격매도라도 단행해야 한다.

# 피봇포인트(pivot-points)

피봇 포인트는 차트 분석가가 추세 변화와 잠재적인 지지/저항 수준을 결정하는데 사용할 수 있는 중요한 지표이다. 피봇 포인트를 통해 이전 기간의 고가,저가 및 종가를 이용하여 향후 기간의 지지 및 저항 수준을 추정할 수 있다.

피봇 포인트를 산출할 때 사용되는 데이터는 차트의 시간간격에 따라 달라진다. 1, 5, 10 및 15분 차트의 경우 전날의 고가, 저가 및 종가를 사용한다. 즉 오늘 일중 차트에 대한 피벗 포인트는 어제의 최고, 최저 및 종가만을 기준한다. 일단 피벗 포인트가 산출되면 변경되지 않고 하루 종일 계속 유지된다.

30,60 및 120분 간격의 차트의 피봇포인트는 이전주의 고가, 저가 및 종가를 사용한다. 산출된 피봇포인트는 일주일 동안 변경되지 않고 유지된다.

일일(일봉) 차트의 피봇포인트는 이전 달 데이터를 사용하여 산출된다. 예를 들어 6월 1일의 피벗포인트는 5월 달의 고가, 저가 및 종가를 기준으로 산출되며 그 값은 6월 한 달 동안 고정된다.

피봇포인트가 산출되는 방식은 다음과 같다.

### $$Pivot\,\,Point\,(P) = \frac{고가 + 저가 + 종가}{3}$$ <br>
### $$지지선\,1\,(S1) = (P \times 2) - 고가$$<br>
### $$지지선\,2\,(S2) = P - (고가 - 저가)$$<br>
### $$저항선\,1\,(R1) = (P \times 2) - 저가$$<br>
### $$저항선\,2\,(R2) = P + (고가 - 저가)$$<br>

## 매매방법

피봇포인트는 분석가에게 가격의 추세를 결정하고 지지 및 저항 수준을 설정할 수 있도록 기준을 제공한다. 피봇포인트를 단독으로 매매에 적용하기 보다는 다른 지표에서 매수/매도 신호가 나왔을 때 주가가 피봇포인트의 지지/저항선에 근접 또는 돌파 하였는지를 판단해 볼 수 있다.

# 모멘텀(momentum)

모멘텀(Momentum)이란 '운동량' 또는 '속도'를 나타내는 물리량으로 주식에서는 주가의 변동폭, 구체적으로 말한다면 n일 전의 종가에 대한 현재의 종가의 차이를 뜻한다. 예컨대 10일간 Momentum이란 10일 전의 종가와 오늘의 종가와의 가격 차이로 산출된다. n일 전 종가에 비하여 현재 종가의 차이가 많다면 그만큼 주가의 속도, 또는 운동량이 많아진 것일 터, 속도가 강화될수록 추세는 더 강해진다. 그러다가 속도가 더 늘어나지 못하고 줄어든다면 추세도 꺾일 것이다. Momentum은 결국 주가의 변동폭, 또는 속도를 구하여 추세의 전환점을 파악하기 위하여 고안된 지표이다.

모멘텀은 다음과 같이 산출된다.

### $$Momentum = price_0 - price_{-n}$$

### $$price_0 = 오늘 \,\,종가$$ <br>
### $$price_{-n} = n 일전 \,\,종가$$ 

## 매매방법

1. Momentum을 살피면 추세가 더 강화되거나 약화되는 모습을 확인할 수 있다. 따라서 Momentum 곡선이 계속 상승하다가 하락세로 돌아서는 순간이 매도 타이밍이고, 반대로 Momentum 곡선이 계속 하락하다가 상승세로 돌아서는 순간이 매입 타이밍이 된다. 하지만 현실에서는 이 방법을 사용하기는 쉽지 않다.
- Momentum의 추세선을 활용하는 방법<br> - 첫 번째 단계로 Momentum의 추세선을 그린다. 이때 Momentum의 추세선은 주식시장의 주가 추세와 동일한 방향이 되어야 한다. <br> - Momentum이 스스로 추세선을 무너뜨리는 시기가 매매 타이밍이다. <br> - 즉 주식시장에서 주가가 상승추세여서 Momentum의 추세선도 상승추세선으로 그려나가다가 Momentum 곡선이 상승추세선을 무너뜨리고 하락하는 순간이 매도 타이밍이 되며, 반대로 주식시장에서 주가가 하락추세여서 Momentum의 추세선도 하락추세선으로 그려나가다가 Momentum 곡선이 하락추세선을 돌파하고 상승하는 순간이 매입 타이밍이 된다. <br> - 만일 주식시장의 추세와 Momentum의 추세선을 서로 다른 방향으로 그려서는 제대로 된 신호를 포착하기 어렵다는 것에 주의하여야 한다.


# ATR(Average True Range) : 평균진정 가격범위

일반저긍로 우리는 어제의 종가와 오늘의 종가와의 차이를 하루 동안의 가격 변동폭이라고 말한다. 그런데 엄밀한 의미로는 그렇게 말하는 것은 정확하지 않다. 왜냐하면 하루 종일 거래가 단지 종가로만 이루어지는 것이 아니기 때문이다. 장중에 주가는 고점도 만들고 저점도 만든다. 따라서 이 모든 것을 다 감안하여 하루 동안의 주가 변동폭을 계산하여야 한다. 이 변동폭을 True Range 라 하고 이를 이동평균한 값을 ATR이라 한다.

ATR을 산출하는 공식은 다음과 같다.

### $$True\,\,High=당일의\,\,고점과\,\,전일의\,\,종가\,\,중에서\,\,높은\,\,것$$ <br>
### $$True\,\,Low=당일의\,\,저점과\,\,전일의\,\,종가\,\,중에서\,\,낮은\,\,것$$ <br>
### $$True\,\,Range=True\,\,High-True\,\,Low$$<br>
### $$ATR=\frac{True\,\,Range + (n-1) \times True\,\,Range_{-1}}{n}$$

n = period: Number(14), 기간

## 매매방법

ATR은 엄밀하게 말하면 매매지표는 아니다. ATR 자체만으로는 매입 또는 매도 신호를 포착하기는 어렵다. 왜냐하면 ATR은 변동성을 측정하는 지표이지 추세의 방향성을 알아보는 지표가 아니기 때문이다. 변동성이 증가하거나 감소하였다고 하여 추세의 방향이 뒤바뀔 것이라고 말할 수는 없다. 따라서 ATR은 독립적으로 사용하기 보다는 다른 지표들의 보조지표로 사용되는 것이 보통이다.

# CMF(Chaikin Money Flow) : 차이킨 자금흐름

주식시장에 유입되거나 또는 주식시장으로부터 빠져나가는 자금의 흐름을 포착하기 위해 개발된 지표. 이 지표를 개발한 마크 차이킨(Marc Chaikin)은 CLV(Close Location Value)를 활용하여 오늘의 주가가 얼마나 오르는지 그 정도를 표현하고자 하였고, 거기에 거래량을 곱하는 방식을 시장의 자금흐름을 포착하려고 노력하였다.

CMF는 다음과 같이 산출된다.

### $$CLV=\frac{(당일종가-당일저가)-(당일고가-당일종가)}{(당일고가-당일저가)}$$ <br>

### $$CMF_i = \frac{\sum_{i-n}^{i}(CLV \times 거래량)}{\sum_{i-n}^{i}거래량}$$

## 매매방법

CMF의 값은 이론적으로 -1 ~ +1 사이의 값을 가지게 되지만 -1 또는 +1이 되는 경우는 거의 없다. 통상 +0.25 또는 -0.25 선이 매매의 기준이 된다.

CMF의 값이 +0.25 아래에서 0.25선을 상향돌파할 때를 매입 타이밍으로 간주하며, 반대로 -0.25 위에서 -0.25 선을 하향돌파할 때를 매도 타이밍으로 간주하는 방법이 보편적으로 사용된다.

# MFI(Money Flow Index) : 자금수급지수

MFI 는 다음과 같은 과정을 거쳐 산출된다.

$$Typical\,\,Price= \frac{당일고가+당일종가+종가}{3}$$

$$Money\,\,Flow=Typical\,\,Price \times 거래량$$

당일의 $Typical\,\,Price \,\, \geq\,$ 전일의 $Typical\,\,Price$ 인 경우 :

> $Positive \,\,Money\,\,Flow=Positive \,\,Money\,\,Flow_{-1} + Money\,\,Flow$

당일의 $Typical\,\,Price \,\, <\,$ 전일의 $Typical\,\,Price$ 인 경우 :

> $Negative \,\,Money\,\,Flow=Negative \,\,Money\,\,Flow_{-1} + Money\,\,Flow$

$$Money\,\,Ratio_i = \frac{\sum_{i-n}^{i}Positive \,\,Money\,\,Flow}{\sum_{i-n}^{i}Negative \,\,Money\,\,Flow}$$

### $$Money\,\,Flow\,\,Index=100 - \,\frac{100}{1 + Money\,\,Ratio}$$

위 계산과정을 '거래량을 고려한 RSI' 라고 생각한다면 MFI를 이해하기 쉽다. MFI 는 RSI의 계산방식에 거래량을 덧붙였다. 왜냐하면 결국 주가의 원동력은 거래량이므로 주식시장의 기술적분석에서 거래량을 무시할 수 없기 때문이다. 따라서 MFI는 RSI보다 더 개선된 지표가 되었다.

이 지표가 사실상 RSI에다 거래량을 덧붙인 것이므로 MFI의 값도 RSI처럼 0에서 100의 범위에서 움직인다.

## 매매방법

MFI의 값이 80 이상이면 현재의 주식시장이 과열국면인 것으로 간주되고, 반대로 MIF의 값이 20 이하를 나타내면 현재의 주식시장이 과매도국면인 것으로 간주된다. 주식시장의 과열국면일 때 매도기회를 노려야 하며, 주식시장이 과매도국면일 때 매입기회를 노려야 하는 것은 당연하다.

> 하지만 다른 지표들의 경우처럼 단지 MFI의 값이 80 이상이라는 이유만으로 매도하고, MFI의 값이 20 이하라는 이유만으로 매입하면 성급한 결정이 될 위험이 높다. 따라서 MFI의 값이 80이상이더라도 일단 느긋하게 기다렸다가 MFI값이 80선을 하향 돌파할 때를 결정적인 매도기회로 삼는 것이 더 효과적이다. 반대로 매입의 경우는 MFI의 값이 20 이하일지라도 일단 느긋하게 기다렸다가 MFI의 값이 20선을 상향돌파 할 때를 결정적인 매입기회로 삼는 것이 더 효과적이다.

# VWAP(Volume Weighted Average Price)

거래량 가중 평균가격(VWAP)는 거래에서 벤치마크로 사용되는 지표로 모든 거래에 대해 거래 된 금액(거래 된 주식수 x 거래량)을 합산한 다음 그날 거래량으로 나누어 계산 된다. 정확한 VWAP는 모든 거래 정보가 담긴 틱 데이터를 기반으로 하지만 모든 틱 데이터를 고려하여 지표를 산출하는것은 매우 큰 리소스를 요구하게 되기에 보통 대푯값이라는 값을 사용해서 계산되어 진다. 대푯값은 어떤 봉이 있을 때 그 봉의 고가, 저가 및 종가의 평균을 의미한다. 따라서 VWAP는 다음과 같은 공식으로 산출되어 진다.

Cumulative(Volume x Typical Price)/Cumulative(Volume)

$$VWAP= \frac{\sum 거래량 \times 대푯값}{\sum 거래량}$$

$$ 대푯값 = \frac{고가 + 저가 + 종가}{3}$$

## 매매방법

VWAP는 매매지표 보다는 거래 효율성을 측정하는 데 사용될 수 있다. 주식을 매매하거나 기관이나 개인이 자신의 매매 가격을 VWAP 가치와 비교할 수 있다. VWAP 가격보다 낮은 가격의 주문은 평균 가격보다 낮은 가격으로 구입했기 때문에 좋은 거래로 간주할 수 있으며 반대로, VWAP보다 높은 가격으로 판매되는 주문은 평균 가격 이상으로 판매 되었기 때문에 좋은 판매로 간주될 수 있다.
